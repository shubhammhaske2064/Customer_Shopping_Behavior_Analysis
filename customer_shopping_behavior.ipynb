```python
# convert all columns names in lower case with attached by underscore
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns={"purchase_amount_(usd)" : "purchase_amount"})
```

```python
df.columns
```

```python
# create a column age_group
labels = ["Young","Adult","Middle_aged","senior"]
df["age_group"] = pd.qcut(df["age"], q=4, labels=labels)
```

```python
df[["age","age_group"]].head()
```

```python
# purchase frequency day 

frequency_mapping = {
    'Fortnightly' : 14, 
    'Weekly' : 7, 
    'Annually' : 360, 
    'Quarterly' : 90, 
    'Bi-Weekly' : 14,
    'Monthly' : 30, 
    'Every 3 Months' : 90
}

df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purchase_frequency_days","frequency_of_purchases", "purchase_frequency_days"]].head(10)
```

```python
# both discount_applied,promo_code_used has same records hence delete one of them

df = df.drop("promo_code_used", axis=1)
df.columns
```
